```python
columnas = ['acciones', 'fechas', 'precios']

df = pd.read_csv(r'D:\Users\USURIO\Desktop\Talento Tech\Notebooks\Matplotlib\dow_jones_index.data',
                header=0,
                usecols=[1,2,6],
                names=columnas,
                parse_dates=['fechas'],
                converters={'precios': lambda x: float(x.replace('$', ''))}
                )
df
```

```python
df.fechas.dt.year.value_counts()
```

#Mes 
#Nombre del mes 
#Columna MES con los número de los meses y Nombre_mes

```python
df['Mes'] = df.fechas.dt.month
df['Mes_nombre'] = df.fechas.dt.month_name()
df
```

El rango en meses de la captura de la información

```python
df.fechas.dt.month_name().value_counts()
```

Los datos se distribuyen desde Enero a Junio del 2011


```python
filtro = df['Mes_nombre'] == 'January'
df[filtro]
```

```python
#groupby 
#Acción por mes y cuántas acciones en el mes

df[filtro].groupby('acciones')['Mes_nombre'].value_counts()
```

## <font color=#003d5c> _¿Cuáles acciones en promedio tuvieron la mayor y la menor cotización?_ </font>

```python
#groupby .agg
#media
x = df.groupby('acciones').agg({'precios': ['mean']}).sort_values([('precios', 'mean')], ascending=False)

x.iloc[[0, -1]]
```

```python
df_new = df[['acciones', 'precios']].groupby('acciones').mean().sort_values(by='precios')
df_new.iloc[[0,-1]]
```

- IBM tuvo el mayor precio: 163,102 USD
- BAC tuvo el menor precio 13,052 USD

## <font color=#003d5c> _¿Cuáles acciones tuvieron la mayor y menor variación de precio?_ </font>

```python
#groupby .agg
#std 

x = df.groupby('acciones').agg({'precios': ['std']}).sort_values([('precios', 'std')], ascending=False)
x.iloc[[0, -1]]
```

- CAT es la acción más volátil, con desviación típica de 6,22 USD
- AA es la acción más consistente, con una desviación típica de 0.77 USD

Crear el DataFrame con las fechas, acciones y los precios que cada acción tuvo en cada fecha

```python
df_nuevo = df.pivot_table(index='fechas', columns='acciones', values='precios')
df_nuevo
```

```python
pv_df = df.pivot_table(index='acciones', columns='fechas', values= 'precios')
pv_df
```

```python
import matplotlib.pyplot as plt

fig, (pan1, pan2) = plt.subplots(2,1,sharex=True, constrained_layout=True)

#CAT
pan1.plot(df_nuevo.index, df_nuevo.CAT, color='blue', marker='o', linestyle='--', label= 'CAT')
#AA
pan2.plot(df_nuevo.index, df_nuevo.AA, color='orange', marker='v', linestyle='solid', label= 'AA')

for pan in fig.get_axes():
    pan.set(xlabel = 'Fechas',
            ylabel = 'Precio USD',
            title = 'Precio de las acciones (cierre)')

pan1.legend()
pan2.legend()
```

## <font color=#003d5c> _¿Existe alguna correlación entre los precios de CAT y el resto de las acciones?_ </font>

```python
df_nuevo.corr().loc['CAT']
```

```python
#Correlación positiva
df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT'] > 0.8]
```

```python
#Correlación negativa
df_nuevo.corr().loc['CAT'][df_nuevo.corr().loc['CAT'] < 0].sort_values(ascending=False)
```
